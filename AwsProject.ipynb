```
# pyspark functions
from pyspark.sql.functions import *
# URL processing
import urllib
```

```
# Define file type
file_type = "csv"
# Whether the file has a header
first_row_is_header = "true"
# Delimiter used in the file
delimiter = ","
# Read the CSV file to spark dataframe
aws_keys_df = spark.read.format(file_type)\
.option("header", first_row_is_header)\
.option("sep", delimiter)\
.load("/FileStore/tables/new_user_credentials.csv")
```

```
# Get the AWS access key and secret key from the spark dataframe
ACCESS_KEY = aws_keys_df.where(col('User name')=='DBricks').select('Access key ID').collect()[0]['Access key ID']
SECRET_KEY = aws_keys_df.where(col('User name')=='DBricks').select('Secret access key').collect()[0]['Secret access key']
# Encode the secrete key
ENCODED_SECRET_KEY = urllib.parse.quote(string=SECRET_KEY, safe='')
```

```
# AWS S3 bucket name
AWS_S3_BUCKET = "dbricksproject"
# Mount name for the bucket
MOUNT_NAME = "/mnt/dbricksproject"
# Source url
SOURCE_URL = "s3n://{0}:{1}@{2}".format(ACCESS_KEY, ENCODED_SECRET_KEY, AWS_S3_BUCKET)
# Mount the drive
dbutils.fs.mount(SOURCE_URL, MOUNT_NAME)
```

```

# File location and type
file_location = "/mnt/dbricksproject/covid_vaccine_statewise.csv"
file_type = "csv"
# CSV options
infer_schema = 'true'
first_row_is_header = 'true'
delimiter = ','
# The applied options are for CSV files. For other file types, these will be ignored.
df = spark.read.format(file_type) \
.option('inferSchema', infer_schema) \
.option('header', first_row_is_header) \
.option('sep', delimiter) \
.load(file_location)
display(df)
```

```
df=df.withColumnRenamed("Updated On","Updated")
df=df.withColumnRenamed("Total Doses Administered","TotalDoses")
df=df.withColumnRenamed("First Dose Administered","FirstDose")
df=df.withColumnRenamed("Second Dose Administered","SecondDose")
df=df.withColumnRenamed("Male (Doses Administered)","MaleDosesAdministered")
df=df.withColumnRenamed("Female (Doses Administered)","FemaleDosesAdministered")
df=df.withColumnRenamed("Transgender (Doses Administered)","TransgenderDosesAdministered")
df=df.withColumnRenamed(" Covaxin (Doses Administered)","CovaxinDosesAdministered")
df=df.withColumnRenamed("CoviShield (Doses Administered)","CoviShieldDosesAdministered")
df=df.withColumnRenamed("Sputnik V (Doses Administered)","SputnikVDosesAdministered")
df=df.withColumnRenamed("18-44 Years (Doses Administered)","18-44YearsDosesAdministered")
df=df.withColumnRenamed("45-60 Years (Doses Administered)","45-60YearsDosesAdministered")
df=df.withColumnRenamed("60+ Years (Doses Administered)","60+YearsDosesAdministered")
df=df.withColumnRenamed("18-44 Years(Individuals Vaccinated)","18-44YearsIndividualsVaccinated")
df=df.withColumnRenamed("45-60 Years(Individuals Vaccinated)","45-60YearsIndividualsVaccinated")
df=df.withColumnRenamed("60+ Years(Individuals Vaccinated)","60+YearsIndividualsVaccinated")
df=df.withColumnRenamed("Male(Individuals Vaccinated)","MaleIndividualsVaccinated")
df=df.withColumnRenamed("Female(Individuals Vaccinated)","FemaleIndividualsVaccinated")
df=df.withColumnRenamed("Transgender(Individuals Vaccinated)","TransgenderIndividualsVaccinate")
df=df.withColumnRenamed("Total Individuals Vaccinated","TotalIndividualsVaccinated")
display(df)






```
